```python
def cuadrado(x):
  return x**2
```

```python
cuadrado(4)
```

```python
lista = list(range(10))
lista
```

```python
lista_2 = []
for numero in lista:
  lista_2.append(cuadrado(numero))

lista_2
```

```python
list(map(cuadrado,lista))
```

list comprehension

```python
lista = list(range(10))
```

```python
lista_2 = [x**2 for x in lista]
lista_2
```

```python
lista_2 = [x**2 for x in lista if x % 2 == 0]
lista_2
```

```python
lista = [(0,0), (1,1), (2,2)]

[ x + y for x,y in lista]
```
