#### Sintaxis para una definición de función

__def__

La sentencia _def_ sirve para crear funciones definidas por el usuario. Una definición de función es una sentencia ejecutable.

~~~
def NOMBRE_FUNCION(PARÁMETROS):
    SENTENCIAS
    return [EXPRESIÓN]
~~~
* __Nombre__: Es el nombre de la función
* __Parámetro__: En las funciones, cuando recibe argumentos, se les denominan parámetros.
* __Sentencias__: Es el bloque del código que ejecuta la función.
* __Return__: Es una sentencia de Python, le indica a la función qué devolver cuando llamamos a la función
* __Expresión__: Es lo que devuelve la funcion.


```python
# Definimos la funcion -- saludar

def saludar():
  print("Hola mundo!")
  return


saludar()
print("Coninua el programa")
```

```python
# definimos la funcion -- despedirse

def despedirse():
  print("Adios!!!")
  return


print("programa principal")
despedirse()
print("continua programa")
```

```python
# definir funciones avanzadas

def saludar_nombre(nombre):
  print(f"Hola {nombre}, bienvenido!!!!")
  return
```

```python
# llamo a la funcion

saludar_nombre("Pablo")
print("Continua el rpgrama")
```

#### Variables y funciones

Vamos a ver el alcance que tiene una variable en python

```python
# Alcance de las variables en las funciones


var_global = "Global"


def var_function():
    var_local = "Local"
    print(f"Adentro de la funcion: var_global: {var_global}")
    print(f"Adentro de la funcion: var_local: {var_local}")
    return



print("Programa principal")

var_function()

print("Continua el programa")

print(f"Afuera de la funcion -- var_global: {var_global}")
```

```python
# Variable local

def test_local():
  nombre_funcion = "Juan"
  print(nombre_funcion)
  return

test_local()
print(nombre_funcion)
```

```python
# modificar una variable Global
nombre = "Juan"

def test():
  global nombre
  nombre = "Pedro"
  print(nombre)
  return


test()
print(nombre)
```

### Momentos de una funcion

Podemos diferenciar dos momentos:

* la definicion de la funcion

* la llamada a la funcion

```python
# definicion
def numero_mas_uno(numero):
  numero += 1
  return numero
```

Durante la definición de la función, las variables o valores se denominan parámetros

```python
# llamada
retorno = numero_mas_uno(5)
print(retorno)
```

Y durante la llamada se denominan argumentos.

```python
# momento.py
def suma_uno(numero):
    numero += 1
    return numero

foo = 60

resultado = suma_uno(foo)

print(f"el numero {foo} mas uno es {resultado}")
```

### Retornando valores

Las funciones pueden comunicarse con el exterior de las mismas, al proceso principal del programa usando la sentencia **return**. La comunicación con el exterior se hace devolviendo valores.


```python
# retornar un valor

def saludar_con_nombre(nombre):
  saludando = f"Hola {nombre}"
  return saludando


mensaje = saludar_con_nombre("Juan")

print(mensaje)

# Que tipo de dato retorna?
```

```python
# perimetro y superficie

def cuadrado():
  lado = 10
  perimetro = lado * 4

  superficie = lado * lado

  return perimetro, superficie

valores = cuadrado()
print(valores)
perimetro_cuadrado, superficie_cuadrado = cuadrado()

print(f"El perimetro del cuadrado es {perimetro_cuadrado}")
print(f"La superficie del cuadrado es {superficie_cuadrado}")
```

#### Media

Escribir una función que reciba una muestra de números en una lista y devuelva su media.

```python
# Definimos la funcion
def media(numeros):
  acumulado = 0

  for numero in numeros:
    acumulado += numero

  promedio = acumulado / len(numeros)

  return promedio

lista_numeros = [4,4,4,4,4,4]

promedio = media(lista_numeros)

print(promedio)
```

#### Multiplo

Crea un programa que le pida dos números enteros al usuario y diga por consola, si alguno de ellos es múltiplo del otro. La función debe llamarse es_multiplo().

```python
# Definicion de la funcion

def es_multiplo(numero1, numero2):
    if (numero1 % numero2 == 0 or numero2 % numero1 == 0):
        return True
    else:
        return False


num1 = 10
num2 = 20
resultado = es_multiplo(num1,num2)
print (f"¿los números {num1} y {num2} son multiplos? {resultado}")

num1 = 15
num2 = 25
resultado = es_multiplo(num1,num2)
print (f"¿los números {num1} y {num2} son multiplos? {resultado}")

num1 = 12
num2 = 4
resultado = es_multiplo(num1,num2)
print (f"¿los números {num1} y {num2} son multiplos? {resultado}")



num1 = int(input("Ingresar un numero: "))
num2 = int(input("Ingresar un numero: "))
resultado = es_multiplo(num1,num2)
print (f"¿los números {num1} y {num2} son multiplos? {resultado}")


```
